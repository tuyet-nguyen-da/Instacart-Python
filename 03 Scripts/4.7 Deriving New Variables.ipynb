# Define a fuction

```python
# define price_label function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low range products'
    elif (row['prices'] >= 5) and (row['prices'] <= 15):
        return 'Mid range products'
    elif row['prices'] > 15:
        return 'High range products'
    else: return np.nan
    
```

```python
# apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# check the output
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# If-Statements with For-Loops
