##### Mixed-type data

```python
# check for mixed type
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# check data types of all columns of the dataframe
df_customers.dtypes
```

```python
df_customers['first_name'].dtype
```

```python
# convert data types of first_name, surname, state to string 
for col in ['first_name', 'surname', 'state'  ]:
    df_customers[col] = df_customers[col].astype('str')
```

```python
# check the result
df_customers['first_name'].dtypes
```

```python
# convert the 'date_joined' column to date data type
df_customers['date_joined'] = pd.to_datetime(df_customers['date_joined'])
```

```python
# check the result
df_customers['date_joined'].dtype
```

```python
# convert 
for col in ['gender', 'fam_status']:
    df_customers[col] = df_customers[col].astype('category')
```

```python
# check the result
df_customers[['gender', 'fam_status']].dtypes
```
